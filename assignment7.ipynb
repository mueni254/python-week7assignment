```python
# DATA VISUALIZATION
# DATA VISUALIZATION
import pandas as pd
import matplotlib.pyplot as plt

# Load data
data = pd.read_csv('Sample7.csv')

# Rename 'data' to 'df' for consistency with the rest of the code
df = data

# Make sure 'Sales till date' is in datetime format
df['Sales till date'] = pd.to_datetime(df['Sales till date'])

# Extract month from the date column
df['Month'] = df['Sales till date'].dt.to_period('M')

# Group by month and calculate total sales
monthly_sales = df.groupby('Month')['Total Sales'].sum().reset_index()

# Group by month and calculate total profit
monthly_profit = df.groupby('Month')['Profit Recorded'].sum().reset_index()

# Merge the two dataframes
monthly_data = pd.merge(monthly_sales, monthly_profit, on='Month')

# Convert Period to string for better plotting
monthly_data['Month'] = monthly_data['Month'].astype(str)

plt.figure(figsize=(12, 6))
plt.plot(monthly_data['Month'], monthly_data['Total Sales'], marker='o', linestyle='-')
plt.plot(monthly_data['Month'], monthly_data['Profit Recorded'], marker='o', linestyle='-')

plt.title('Total Sales by Month/profit')
# If plotting profit: plt.title('Total Profit by Month')
plt.xlabel('Month')
plt.ylabel('$ Sales/profit')
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()


```

```python
#BAR CHART

# BAR CHART TO MAP TOTAL SALES PER REGION
import pandas as pd
import matplotlib.pyplot as plt



# Assuming df is your dataframe and it has columns 'Region' and 'Total Sales'
# Group by region and calculate total sales
State_sales = df.groupby('State')['Total Sales'].sum().reset_index()

# Sort by sales value for better visualization
State_sales = State_sales.sort_values('Total Sales', ascending=False)

# Create a bar chart
plt.figure(figsize=(12, 6))
bars = plt.bar(State_sales['State'], State_sales['Total Sales'], color='skyblue')

# Add labels and title
plt.title('Total Sales by State', fontsize=16)
plt.xlabel('State', fontsize=12)
plt.ylabel('Total Sales ($)', fontsize=12)
plt.xticks(rotation=45, ha='right')  # Rotate labels for better readability

# Add data labels on top of bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 0.1,
             f'${height:,.0f}', ha='center', fontsize=10)

plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

```

```python
# HISTOGRAM FOR TOTAL SALES

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Assuming df is your dataframe with the column 'Total Sales'
# Create a histogram
plt.figure(figsize=(10, 6))

# Create histogram with customized bins
# You may need to adjust the bins based on your actual data distribution
bins = np.linspace(df['Total Sales'].min(), df['Total Sales'].max(), 15)  # 15 bins
plt.hist(df['Total Sales'], bins=bins, color='skyblue', edgecolor='black', alpha=0.7)

# Add labels and title
plt.title('Distribution of Total Sales', fontsize=16)
plt.xlabel('Total Sales ($)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

# Add grid for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Format x-axis with dollar signs
plt.gca().xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'${x:,.0f}'))

# Add mean line
mean_sales = df['Total Sales'].mean()
plt.axvline(mean_sales, color='red', linestyle='dashed', linewidth=1)
plt.text(mean_sales*1.02, plt.ylim()[1]*0.9, f'Mean: ${mean_sales:,.2f}', 
         color='red', fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Prepare Data: Group sales by state and sum
state_sales = df.groupby('State')['Total Sales'].sum().reset_index()

# 2. Sort data (optional: show largest slices first)
state_sales = state_sales.sort_values('Total Sales', ascending=False)

# 3. Create Pie Chart
plt.figure(figsize=(10, 8))

# Custom colors (optional)
colors = plt.cm.tab20.colors[:len(state_sales)]

# Plot with autopct to show percentages
plt.pie(
    state_sales['Total Sales'],
    labels=state_sales['State'],
    autopct='%1.1f%%',  # Show percentages with 1 decimal
    startangle=90,      # Rotate to align largest slice at top
    colors=colors,
    wedgeprops={'edgecolor': 'white', 'linewidth': 0.5},  # Add white borders
    textprops={'fontsize': 10}
)

# 4. Add Title and Legend
plt.title('Total Sales Distribution by State', fontsize=16, pad=20)

# Legend with state names and sales values (for clarity)
legend_labels = [f"{state}: ${sales:,.0f}" 
                for state, sales in zip(state_sales['State'], state_sales['Total Sales'])]
plt.legend(
    legend_labels,
    title='States',
    bbox_to_anchor=(1, 0.5),  # Place legend outside the chart
    loc='center left'
)

# 5. Equal Aspect Ratio (to avoid oval-shaped pie)
plt.axis('equal')

plt.tight_layout()
plt.show()
```
